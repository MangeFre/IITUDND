# THE CELL BELOW CONTAINS THE FILENAME CONSTANTS THAT NEED TO BE CHANGED

```python

# this file is a line seperated list of retrieved tweet JSONs to check for active images
INPUT_JSON_FILE = os.getcwd() + '/../data/irma_extras.json'

# this file is a line seperated list of JSONs, the script extracts the set of users from this count images per user
USER_ID_FILE = os.getcwd() + "/../data/CrisisMMD_v1.0/json/hurricane_irma_final_data.json"

# the file to write the retrieved tweets too
OUTFILE = "irma_pic_tweets.json"
```

```python
# Magnus's func for detecting active links
def exists(path):
  r = requests.head(path)
  return r.status_code == requests.codes.ok
```

```python
# get set of user id's to filter by
user_ids = set()  
with open(USER_ID_FILE) as fin:
    for line in fin:
        json_data = json.loads(line)
        user_id = json_data['user']['id']
        user_ids.add(user_id)
```

```python
# get the counts (per user) of images of tweets associated with the retrieved tweets
user_image_counts = dict.fromkeys(user_ids,0)

media_total = 0 # total number of tweets with some kind of media (ie video, pic, etc)
pic_total = 0 # total number of tweets with pictures
count_active = 0 # total number with living links

# for each tweet see if it has media that is a picture and if it is active
with open(INPUT_JSON_FILE) as fin:
    
    # load a tweet JSON
    for line in fin:
        json_data = json.loads(line)
        
        # check if it has media
        if 'extended_entities' in json_data:
            for medium in json_data['extended_entities']['media']:
                media_total += 1
                print('media_total #', media_total)
                
                # check if the media is a pic
                if medium['type'] == 'photo':
                    pic_total += 1
                    print('pic_total #', pic_total)
                    
                    # check if the pic has an active link
                    if exists(medium['media_url_https']):
                        count_active += 1
                        user_id = json_data['user']['id'] 
                        user_image_counts[user_id] = user_image_counts.get(user_id, 0) + 1
                        print('count_active #', count_active)
                        
                        # save this json to outfile
                        with open(OUTFILE, 'a') as fout:
                            fout.write(line)
                    
print('done!')


# get user image history length info (key will be length, val is number of users with that length)
freq = {}
for val in user_image_counts.values():
    freq[val] = freq.get(val, 0) + 1
freq

```
